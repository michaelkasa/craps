```python
import sys
sys.path.append('C:\\Users\\Michael\\OneDrive\\Documents\\Python_programs\\craps')
from craps import *
from numpy import mean
from numpy import array
from numpy import size
from numpy import vstack, hstack, transpose

import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
''' Run simulation 10000 times '''
mike_winnings = []
hooman_winnings = []
N_runs = 10000

''' Get Mike's results '''
for ii in range(N_runs):
    ''' Bet pass always, and quit after losing $60 or playing 30 rounds '''
    mike = Player(bets_pass, quits_after_gainG_or_lossL_or_roundMax)
    mike.gainG = 999999999999999999
    mike.lossL = 70
    mike.roundMax = 30
    board = Board()
    while not mike.is_quitting():
        board.reset()
        while not board.get_status().round_is_over:
            board.take_bets(mike.make_bets(board.get_status()))
            r = board.roll()
        mike.get_payouts(board.return_payouts())
        
    ''' Append the winnings to the array '''
    mike_winnings.append(mike.winnings)
    
''' Get Hooman's results '''
for ii in range(N_runs):
    ''' Bet pass always, and quit after losing $60 or playing 30 rounds '''
    hooman = Player(bets_pass_and_odds, quits_after_gainG_or_lossL_or_roundMax)
    hooman.gainG = 999999999999999999
    hooman.lossL = 70
    hooman.roundMax = 30
    board = Board()
    while not hooman.is_quitting():
        board.reset()
        while not board.get_status().round_is_over:
            board.take_bets(hooman.make_bets(board.get_status()))
            r = board.roll()
        hooman.get_payouts(board.return_payouts())
        
    ''' Append the winnings to the array '''
    hooman_winnings.append(hooman.winnings)
```

```python
''' Put winnings in a 2 x 10000 numpy array '''
arr_winnings = transpose(vstack([array(mike_winnings),
                                 array(hooman_winnings)]))

''' Plot histogram '''
n, bins, patches = plt.hist(arr_winnings, bins=25, 
                            color=['purple','burlywood'],
                            label=['Mike','Hooman'])
plt.xlabel('Winnings (profit in $)')
plt.ylabel('Number of simulations (10,000 total)')
plt.title('Winning probabilities')
plt.grid(True)
plt.axis([-80,160,0,3000])
plt.legend()
plt.show()

''' Plot cumulative histogram '''
n, bins, patches = plt.hist(arr_winnings, bins=25, cumulative=True,
                            color=['purple','burlywood'],
                            label=['Mike','Hooman'])
plt.xlabel('Winnings (profit in $)')
plt.ylabel('Number of simulations (10,000 total)')
plt.title('Cumulative sum of winning probabilities')
plt.grid(True)
plt.axis([-80,160,0,10000])
plt.legend()
plt.show()

print("Mike's average  : " + str(mean(mike_winnings)))
print("Hooman's average: " + str(mean(hooman_winnings)))
```
